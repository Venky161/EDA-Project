```python
Brand = []
for i in df["Title"]:
    r = re.findall(r"^\w+",i)
    if r:
        Brand.append(r[0])
    else:
        Brand.append(np.nan)
        
```

```python
df["Brand"] = df["Title"].apply(lambda x : re.findall(r"^\w+",x)[0] if re.findall(r"^\w+",x) else np.nan)
```

```python
df["Model"] = df["Title"].apply(lambda x : re.findall(r"^\w+\s(.+)\s\(",x)[0] if re.findall(r"^\w+\s(.+)\s\(",x) else np.nan)
```

```python
df["Color"] = df["Title"].apply(lambda x : re.findall(r"\((\w+\s?\w+),",x)[0] if re.findall(r"\((\w+\s?\w+),",x) else np.nan)
```

```python
df["Strage"] = df["Title"].apply(lambda x : re.findall(r",\s(\d+)\sGB",x)[0] if re.findall(r",\s(\d+)\sGB",x) else np.nan)
```

```python
df.drop("Title",axis = 1,inplace = True)
```

```python
df[["Rating","Reviews"]] =df["Reviews and Rating"].str.split("&",expand = True)
```

```python
df.drop("Reviews and Rating",axis = 1,inplace = True)
```

```python
df["Front_Primary_Camera"] = df["Camera"].str.split("|",expand = True).iloc[:,1].str.split("+",expand = True).iloc[:,0]
```

```python
df[["Rear1","Rear2"]] = df["Camera"].str.split("|",expand = True).iloc[:,0].str.split("+",expand = True).iloc[:,0:2]
```

```python
df.drop("Camera",axis = 1, inplace =True)
```
